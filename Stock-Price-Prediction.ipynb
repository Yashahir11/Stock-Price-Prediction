# 9. Time series based Prediction

####  Time Series Stock Price Prediction

In the above sections, a simple machine learning approach was used for stock price prediction.
However, stock prices are time-dependent, meaning current prices are influenced by past values.

To handle this, a Time Series based approach is implemented below using:
- Date as index
- Lag features
- Rolling mean features

This allows the model to capture historical trends more effectively.


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv("ADANIPORTS.csv")
df.head()
```

```python
df = df.drop(columns=['Symbol', 'Series'])
```

```python
df_ts = df.copy()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Trades'])
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df = df.reset_index(drop=True)

df['Date'] = pd.date_range(
    start='2015-01-01',   
    periods=len(df),
    freq='B'             
)

df.set_index('Date', inplace=True)
```

```python
df.head()

```

```python
df_ts = df[['VWAP']].copy()

df_ts['lag_1'] = df_ts['VWAP'].shift(1)
df_ts['lag_2'] = df_ts['VWAP'].shift(2)
df_ts['rolling_mean_5'] = df_ts['VWAP'].rolling(5).mean()
df_ts['rolling_std_5'] = df_ts['VWAP'].rolling(5).std()

df_ts.dropna(inplace=True)
```

```python
train_size = int(len(df_ts) * 0.8)

train = df_ts.iloc[:train_size]
test  = df_ts.iloc[train_size:]

X_train = train.drop('VWAP', axis=1)
y_train = train['VWAP']

X_test = test.drop('VWAP', axis=1)
y_test = test['VWAP']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

model = RandomForestRegressor(
    n_estimators=300,
    random_state=42
)

model.fit(X_train, y_train)
pred = model.predict(X_test)

print("MAE :", mean_absolute_error(y_test, pred))
print("RMSE:", mean_squared_error(y_test, pred, squared=False))
print("R2  :", r2_score(y_test, pred))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(y_test.index, y_test.values, label='Actual VWAP')
plt.plot(y_test.index, pred, label='Predicted VWAP')
plt.legend()
plt.title("VWAP Time Series Prediction (Random Forest)")
plt.xlabel("Date")
plt.ylabel("VWAP")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)

pred_lr = lr.predict(X_test)
print("LR R2:", r2_score(y_test, pred_lr))
```

```python
from sklearn.model_selection import TimeSeriesSplit

tscv = TimeSeriesSplit(n_splits=5)
r2_scores = []

for train_idx, test_idx in tscv.split(df_ts):
    train, test = df_ts.iloc[train_idx], df_ts.iloc[test_idx]
    
    X_train = train.drop('VWAP', axis=1)
    y_train = train['VWAP']
    X_test = test.drop('VWAP', axis=1)
    y_test = test['VWAP']
    
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    r2_scores.append(r2_score(y_test, preds))

print("CV R2 scores:", r2_scores)
print("Mean R2:", sum(r2_scores)/len(r2_scores))
```
